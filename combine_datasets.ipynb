```python
#Import pandas
import pandas as pd
```

```python
#Read in 3 csv files
hof = pd.read_csv('NBA HOF Players.csv')
borderline = pd.read_csv('Borderline HOF Players.csv')
roleplayers = pd.read_csv('NBA Role Players.csv')

#Display
display(hof.head())
display(borderline.head())
display(roleplayers.head())
```

```python
#Subset borderline to fix mistake
borderline = borderline[['Name', 'Position', 'Games', 'Career Length',
       'PPG', 'RPG', 'APG', 'PER', 'FG%', 'FT%', 'Win Shares', 'All-Stars',
       'All-NBA', 'All-Defense', 'All-Rookie Team', 'MVPs', 'Chips', 'ROY',
       'DPOYs', 'Scoring Champ', 'HOF']]
borderline.head()
```

```python
#Check shape for each
print(f"Number of rows in HOF dataset: {hof.shape[0]}, number of columns: {hof.shape[1]}")
print(f"Number of rows in borderline dataset: {borderline.shape[0]}, number of columns: {borderline.shape[1]}")
print(f"Number of rows in role players dataset: {roleplayers.shape[0]}, number of columns: {roleplayers.shape[1]}")
```

Great all the datasets have the same number of columns. 

```python
#Check total number of Non-HOF Players
print(f"Total number of Non-HOF Players: {borderline.shape[0] + roleplayers.shape[0]}, not consideration potential duplicates")
```

```python
#Concat Non-hof dataset
non_hof = pd.concat([borderline, roleplayers], ignore_index=True)

#Check head
non_hof.head()
```

```python
#Check shape
non_hof.shape
```

```python
#Check Non-HOF player columns
non_hof.columns
```

```python
#Check HOF player columns
hof.columns
```

```python
#Convert all column names to lowercase
non_hof.columns = non_hof.columns.str.lower()

#Remove leading and lagging white spaces just in case they exist
non_hof.columns = non_hof.columns.str.strip()

#Remove % and replace with nothing
non_hof.columns = non_hof.columns.str.replace('%', '')

#Replace dashes with spaces
non_hof.columns = non_hof.columns.str.replace('-', ' ')

#Replace spaces with underscores
non_hof.columns = non_hof.columns.str.replace(' ', '_')

#Rename all_rookie_team to rookie team
non_hof.rename(columns={
    'all_rookie_team': 'rookie_team'
},inplace=True)

#Print out new column names
non_hof.columns
```

```python
#Concat both players into a single dataframe
df = pd.concat([hof, non_hof], ignore_index=True)

#View head
df.head()
```

```python
#Check shape
df.shape
```

```python
#Convert Carmelo Anthony to HOF since he was recently abducted
df.loc[152, 'hof'] = 1

#Check result
df.loc[[152]]
```

```python
#Check for potential duplicate values
duplicates = df.duplicated()
duplicates.value_counts()
```

```python
#Check duplicate rows
df[duplicates]
```

```python
#Drop duplicates
df = df.drop_duplicates()
```

```python
#Save to csv file
df.to_csv('NBA Players.csv', index=False)
```

```python
#Subset data to seperate HOF players and Non-HOF players
hof_players = df[df['hof'] == 1]
non_hof_players = df[df['hof'] == 0]

#Check result
print(hof_players['hof'].unique())
print(non_hof_players['hof'].unique())
```

```python
#Take a random sample from each dataframe using the sample() method
hof_sample = hof_players.sample(n=50, replace=False, random_state=42, ignore_index=True)
nonhof_sample = non_hof_players.sample(n=50, replace=False, random_state=42, ignore_index=True)
```

```python
#Concat both dataframes
sds_320_df = pd.concat([hof_sample, nonhof_sample], ignore_index=True)

#Check head and tail
display(sds_320_df.head())
display(sds_320_df.tail())
```

```python
#Check value counts to ensure 50/50 split for hof status
sds_320_df['hof'].value_counts()
```

```python
#Check which players were selected
sds_320_df['name'].unique()
```

```python
#Subset data to include only the features we want
final_df = sds_320_df[['name', 'ppg', 'hof', 'career_length']]

#View head
final_df.head()
```

```python
#Set the index to name
final_df.set_index('name', inplace=True)

#View result
final_df.head()
```

```python
#Save to dataframe
final_df.to_csv('nba_analysis_sds320e.csv',index=True)
```
